```python
target=['Tshirt','Trousers','pullover','dress','coat','sandal','shirt','sneakers','bag','ankleboot']
```

```python
x_train=train_df.drop(['label'],axis='columns').to_numpy()
x_train=x_train.reshape(-1,28,28,1)
x_train=x_train/255
y_train=train_df['label']
```

```python
x_test=test_df.drop(['label'],axis='columns').to_numpy()
x_test=x_test.reshape(-1,28,28,1)
x_test=x_test/255
y_test=test_df['label']
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
for i in range (25):
  plt.subplot(5,5,i+1)
  plt.xticks([])
  plt.yticks([])
  plt.imshow(x_train[i],cmap='Greys')
  plt.xlabel(target[y_train[i]])
```

```python
from keras.models import Sequential
from keras.layers import Conv2D,MaxPool2D,Dropout,Flatten,Dense
```

```python
model=Sequential()
```

```python
model.add(Conv2D(filters=64,kernel_size=(3,3),input_shape=(28,28,1),activation='relu'))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Dropout(rate=0.3))
model.add(Flatten())
model.add(Dense(32,activation='relu'))
model.add(Dense(10,activation='sigmoid'))
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
model.summary()
```

```python
model.fit(x_train,y_train,epochs=50,validation_split=0.05,batch_size=1500)
```

```python
model.evaluate(x_test,y_test)
```
